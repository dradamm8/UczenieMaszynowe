```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("titanic.csv", header = 0, index_col = 0)
df.replace({'CabinReduced': {'n': np.nan}}, inplace = True)
df
```

Funkcja <b>train_test_split</b> dzieli zbiÃ³r danych (przekazany jako lista, numpy array lub DataFrame) na zbiÃ³r treningowy i testowy.
Jako parametry moÅ¼na podaÄ‡ jaki odsetek caÅ‚ego zbioru ma zostaÄ‡ uwzglÄ™dniony przy tworzeniu zbioru (treningowego lub testowego).
JeÅ›li ktÃ³raÅ› z wartoÅ›ci nie jest podana, to domyÅ›lnie odsetek drugiego jest wyliczany jako dopeÅ‚nienie do 1 odsetka pierwszego zbioru.

DomyÅ›lnie (bez podania Å¼adnej z tych wielkoÅ›ci): ZbiÃ³r treningowy - 75%, zbiÃ³r testowy - 25%

Funkcja zwraca listÄ™ podzielonych zbiorÃ³w

```python
col_name = ['cabin', 'CabinReduced', 'sex']
```

```python
X = df[col_name]
y = df[['survived']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
names = ['X_train', 'X_test', 'y_train', 'y_test']
var = [X_train, X_test, y_train, y_test]

for i in range(4):
    print(f"Rozmiary zbioru {names[i]}: {var[i].shape}")
```

Zbiory treningowe majÄ… po 916 wierszy, a testowe - 393. Stanowi to odpowiednio 70% i 30% caÅ‚ego zbioru.
Suma iloÅ›ci wierszy w zbiorach treningowych i testowych jest rÃ³wna iloÅ›ci rekordÃ³w w caÅ‚ym zbiorze danych.

<b>Sprawdzamy czy rozkÅ‚ad etykiet jest rÃ³wnomierny</b>

```python
print("IloÅ›Ä‡ unikalnych etykiet w zbiorze testowym, ktÃ³re nie wystÄ™pujÄ… w treningowym")
for name in col_name:
    Unique_test = [x for x in X_test[name].unique() if x not in X_train[name].unique()]
    print(f"{name}: {len(Unique_test)}")
```

Aby wyciÄ…gnÄ…Ä‡ dobre wnioski, sprawdÅºmy iloÅ›Ä‡ unikalnych etykiet dla danych zmiennych (w zaleÅ¼nosci od tego czy jest to zbiÃ³r testowy czy treningowy)

```python
for name in col_name:
    print(f"IloÅ›Ä‡ unikalnych etykiet dla zmiennej {name}:")
    print(f"Dla zbioru testowego: {len(X_test[name].unique())}")
    print(f"Dla zbioru treningowego: {len(X_train[name].unique())}")
    print(f"Dla caÅ‚ego zbioru: {len(df[name].unique())}")
    print()
```

Dla zmiennej cabin w zasadzie poÅ‚owa etykiet, ktÃ³re wystÄ™pujÄ… w zbiorze testowym, nie wystÄ™pujÄ… w zbiorze treningowym.
Oznacza to, Å¼e rozkÅ‚ad moÅ¼e byÄ‡ zaburzony i, o ile zmienna cabin bÄ™dzie kluczowa przy dalszej analizie, model moÅ¼e daÄ‡ sÅ‚absze wyniki dla zbioru testowego (nie bÄ™dzie moÅ¼liwoÅ›ci odpowiedniego uwzglÄ™dnienia zmiennych w trakcie uczenia). 

JednakÅ¼e przed tym uchroniÄ‡ nas moÅ¼e redukcja, gdzie iloÅ›Ä‡ etykiet zostaÅ‚a drastycznie pomniejszona, a rozkÅ‚ad w zbiorze testowym i treningowym jest juÅ¼ rÃ³wnomierny (CabinReduced).

<b>Mapowanie zmiennych:</b>

```python
#tablica sÅ‚ownikÃ³w do mapowania
dicts = []

for name in col_name:
    d = dict((name, count) for count, name in enumerate(df[name].unique()))
    dicts.append(d)
    
```

```python
new_name = ['cabin_map', "cabin_reduced_map", "sex_map"]

X_train2 = pd.DataFrame()
X_test2 = pd.DataFrame()

for i in range(3):
    X_test2[new_name[i]] = X_test[col_name[i]].map(dicts[i], na_action = 'ignore').values
    X_train2[new_name[i]] = X_train[col_name[i]].map(dicts[i], na_action = 'ignore').values

X_test2
```

<b>Liczba brakujÄ…cych wartoÅ›ci dla nowych etykiet:</b>

```python
print("Dla zbioru testowego:")
print(X_test2[new_name].isnull().sum())
print()
print("Dla zbioru treningowego:")
print(X_train2[new_name].isnull().sum())
```

PorÃ³wnajmy to z sytuacjÄ… przed mapowaniem:

```python
print("Dla zbioru testowego:")
print(X_test[col_name].isnull().sum())
print()
print("Dla zbioru treningowego:")
print(X_train[col_name].isnull().sum())
```

A zatem zmapowanie etykiet nie dotyczyÅ‚o wartoÅ›ci NaN. <br>
Na numeryczne zostaÅ‚y zmapowane tylko wartoÅ›ci niebrakujÄ…ce.

<b>Zmiana wartoÅ›ci brakujÄ…cych na 0</b>

```python
for name in new_name:
    X_test2.replace({np.nan : 0}, inplace = True)
    X_train2.replace({np.nan : 0}, inplace = True)

X_test2
```

Takie dziaÅ‚anie moÅ¼e jednak nie byÄ‡ najlepszym wyjÅ›ciem, poniewaÅ¼ 0 zostaÅ‚o juÅ¼ wykorzystane do mapowania i oznaczaÅ‚o innÄ… wartoÅ›Ä‡.
Wobec tego moÅ¼na postÄ…piÄ‡ inaczej: mapowanie (przy uÅ¼yciu funkcji enumerate) wykonaÄ‡ <b>zaczynajÄ…c liczenie od 1</b>.
Wtedy zmapowanie wartoÅ›ci NaN na 0 bÄ™dzie jednoznaczne.

```python
#tablica sÅ‚ownikÃ³w do mapowania
dicts = []

for name in col_name:
    d = dict((name, count) for count, name in enumerate(df[name].unique(), 1))
    dicts.append(d)

#mapowanie
for i in range(3):
    X_test2[new_name[i]] = X_test[col_name[i]].map(dicts[i], na_action = 'ignore').values
    X_train2[new_name[i]] = X_train[col_name[i]].map(dicts[i], na_action = 'ignore').values

    #zmiana wartoÅ›ci brakujÄ…cych na 0
    X_test2.replace({np.nan : 0}, inplace = True)
    X_train2.replace({np.nan : 0}, inplace = True)
    
X_train2
```

<b>PorÃ³wnanie iloÅ›ci unikalnych etykiet przed i po mapowaniu:</b>

```python
for i in range(3):
    
    name = col_name[i]
    map_name = new_name[i]

    print(f"IloÅ›Ä‡ unikalnych etykiet dla zmiennej {name}")
    print("Przed mapowaniem:")
    print(f"ZbiÃ³r testowy: {len(X_test[name].unique())}")
    print(f"ZbiÃ³r treningowy: {len(X_train[name].unique())}")
    print()
    print("Po mapowaniu:")
    print(f"ZbiÃ³r testowy: {len(X_test2[map_name].unique())}")
    print(f"ZbiÃ³r treningowy: {len(X_train2[map_name].unique())}")
    print('====================================\n')
    
```

Jak widaÄ‡, wartoÅ›ci przed i po mapowaniu nie ulegÅ‚y zmianie.
UdaÅ‚o siÄ™ zatem zakodowaÄ‡ etykiety jako wartoÅ›ci liczbowe, dziÄ™ki czemu bÄ™dÄ… mogÅ‚y zostaÄ‡ uÅ¼yte przy tworzeniu modelu.

BiorÄ…c pod uwagÄ™, Å¼e wartoÅ›ci zostaÅ‚y zakodowane jednoznacznie, to mapowanie nie zmieni to wpÅ‚ywu na jakoÅ›Ä‡ modelu.
WpÅ‚yw moÅ¼e mieÄ‡ za to redukcja zmiennej cabin, gdzie iloÅ›Ä‡ kategorii zostaÅ‚a drastycznie zmniejszona.
Redukcja tej zmiennej zapewniÅ‚a nam takÅ¼e to, Å¼e rozkÅ‚ad etykiet w zbiorze testowym i treningowym bÄ™dzie rÃ³wnomiernym, czego nie udaÅ‚o siÄ™ osiÄ…gnÄ…Ä‡ przed redukcjÄ….
